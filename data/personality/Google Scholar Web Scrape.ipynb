```python
url = 'https://scholar.google.com/scholar?hl=en&as_sdt=0%2C48&q=big+five&btnG='
browser = Chrome()
browser.get(url)
html = browser.page_source
# mc = pymongo.MongoClient()
# coll = mc['big_five_papers']
# big_five_articles = db['articles']
```

```python
soup = BeautifulSoup(html, 'html.parser')
tweets = soup.select('div.tweet')
```

```python
def parse_tweet(tweet):
    username = tweet.select_one('span.username').text
    fullname = tweet.select_one('strong.fullname').text
    text = tweet.select_one('p.tweet-text').text
    tweet_time = tweet.select_one('span._timestamp').attrs['data-time']
    time = pd.Timestamp.fromtimestamp(int(tweet_time))
    tweet_dict = {
        'username': username,
        'fullname': fullname,
        'text': text,
        'tweet_timestamp': time
    }
    return tweet_dict
```

```python
parsed_tweet = [parse_tweet(tweet) for tweet in tweets]
pd.DataFrame(parsed_tweet)
```
